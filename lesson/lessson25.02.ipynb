```python
"""скачать викидамп, create structure for data, visualize structure in matplotlib"""

import re
```

```python
#dump extracted through WikiExtractor.py from command line
with open(r"E:\Lorenzo Tosi\OneDrive - НИУ Высшая школа экономики\Documenti\Program\OOBOONTOO\Practical_1_Wiki\wiki.txt", "r", encoding="utf-8") as r:
    articles = r.read()
    print(articles[:5000])
```

```python
#preprocessing
articles = articles.split("\n\n")
for i, article in enumerate(articles):
        articles[i] = re.sub("\\n", " ", article)
articles[0] = articles[0].lstrip()
```

```python
class WikiArticle():
    def __init__(self, title = "", content = ""):
        self.title = title
        self.content = content
    def parametercalc(self):
        length = len(content)
        wordsintitle = len(title.split())
        if len(content) != 0:
            startswith = content[0]
            endswith = content[-1]
        else:
            startswith = ""
            endswith = ""
        return (length, wordsintitle, startswith, endswith)
        
```

```python
for i, article in enumerate(articles):
    index = article.find(":")
    title = article[0:index]
    content = article[index+2:]
    articles[i] = WikiArticle(title, content)
```

```python
plotlist = []
for i in range(len(articles)):
    plotlist.append(articles[i].parametercalc())
```
